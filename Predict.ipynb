```python
# Importing Ridge regression model from scikit-learn
# Creating a Ridge regression model with regularization parameter alpha=0.1
from sklearn.linear_model import Ridge

rr = Ridge(alpha=0.1)
```

```python
# Selecting predictor columns by excluding 'target', 'name', and 'station' columns from the 'weather' DataFrame
predictors = weather.columns[~weather.columns.isin(["target", "name", "station"])]
```

```python
predictors
```

```python
# Perform backtesting of the model's predictions using a rolling window approach
def backtest(weather, model, predictors, start=3650, step=90):
    all_predictions = []

    for i in range(start, weather.shape[0], step):
        train = weather.iloc[:i,:]
        test = weather.iloc[i:(i+step),:]
        
        model.fit(train[predictors], train["target"])
        
        preds = model.predict(test[predictors])
        
        combined = pd.concat([test["target"], pd.Series(preds, index=test.index)], axis=1)
        
        combined.columns = ["actual", "prediction"]
        combined["diff"] = (combined["prediction"] - combined["actual"]).abs()
        
        all_predictions.append(combined)

    return pd.concat(all_predictions)
```

```python
predictions = backtest(weather, rr, predictors)
```

```python
predictions
```

```python
# Calculate the mean absolute error between actual and predicted values
from sklearn.metrics import mean_absolute_error

mean_absolute_error(predictions["actual"], predictions["prediction"])
```

```python
# Define a function to compute rolling averages and percentage differences
def pct_diff(old, new):
    return (new - old)/old
def compute_rolling(weather, horizon, col):
    label = f"rolling_{horizon}_{col}"
    
    weather[label] = weather[col].rolling (horizon).mean()
    weather[f"{label}_pct"] = pct_diff(weather[label], weather[col])
    return weather
    
rolling_horizons = [3, 14]
                    
for horizon in rolling_horizons: 
    for col in ["tmax", "tmin","prcp"]:
        weather = compute_rolling(weather, horizon, col)
```

```python
# Remove the first 14 days of data due to lack of previous days' data
weather = weather.iloc[14:,:]
```

```python
# Fill missing values in the weather DataFrame with zeros
weather = weather.fillna(0)
```

```python
# Calculate expanding mean for each month and day of the year for temperature and precipitation columns
def expand_mean(df) :
    return df.expanding(1).mean()
for col in ["tmax", "tmin", "prcp"]:
    weather[f"month_avg_{col}"] = weather[col].groupby(weather.index.month, group_keys=False).apply(expand_mean)
    weather[f"day_avg_{col}"] = weather[col].groupby(weather.index.day_of_year, group_keys=False).apply(expand_mean)
```
